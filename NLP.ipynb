```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
```

```python
nltk.download('punkt') #NLTK, or the Natural Language Toolkit, is a powerful library in 
                        #Python used for working with human language data (text).
nltk.download('stopwords')
```

```python
# Text cleaning function
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\b\w{1,2}\b', '', text)  # Remove short words
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text
```

```python
# Apply text cleaning
df['Cleaned_Comment'] = df['Comment'].apply(clean_text)
```

```python
# Display the cleaned comments
df[['Comment', 'Cleaned_Comment']]
```

```python
# Tokenization and stopword removal
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
def tokenize_and_remove_stopwords(text):
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```python

df['Processed_Comment'] = df['Cleaned_Comment'].apply(tokenize_and_remove_stopwords)
```

                       Preprocessing Techniques Used and Their Impact on Model Performance
Lowercasing

What: Converts all characters to lowercase.
Why: Standardizes the text, reducing the number of unique words (e.g., "Hello" and "hello" become the same).
Impact: Improves efficiency by ensuring different cases of the same word are treated equally.

Removal of Short Words

What: Removes words that are one or two characters long.
Why: Short words often carry less meaningful information.
Impact: Reduces noise, focusing on more informative words, enhancing model performance.

Removal of Punctuation

What: Removes punctuation marks.
Why: Punctuation usually doesn't carry significant meaning in text analysis.
Impact: Simplifies text, leading to better feature extraction.

Tokenization and Stopword Removal

What: Splits text into individual words (tokens) and removes common stopwords (e.g., "the", "is").
Why: Tokenization breaks down text into manageable pieces, while stopwords often carry little semantic information.
Impact: Reduces dimensionality and noise, focusing on more informative words.

```python
# Initialize CountVectorizer
count_vectorizer = CountVectorizer()
```

```python
# Fit and transform the text data
X_count = count_vectorizer.fit_transform(df['Cleaned_Comment'])
```

```python
# Display the first few rows of the feature matrix
print(X_count_df.head())
```

CountVectorizer creates a matrix of token counts for each document.
How It Works: It tokenizes the text and counts the occurrences of each word. Each row in the resulting matrix corresponds to a document, and each column corresponds to a unique word in the corpus. The cell values represent the count of each word in the respective document.
Impact: This method transforms text into a numerical format where the frequency of each word is represented. While straightforward, it doesn’t account for the relative importance of words across documents or handle synonyms and word variations well.

TfidfVectorizer calculates a matrix where each word is weighted based on its term frequency (how often it appears in a document) and inverse document frequency (how common it is across all documents).
How It Works: It first counts the occurrences of each word (like CountVectorizer) but then applies a weighting formula. The formula decreases the weight of common words and increases the weight of rare words, providing a measure of word importance in each document.

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Initialize Naive Bayes model
nb_model = MultinomialNB()
```

```python
# Train the model
nb_model.fit(X_train, y_train)
```

```python
# Predict on the test set
nb_predictions = nb_model.predict(X_test)
```

```python
# Evaluate the model
nb_accuracy = accuracy_score(y_test, nb_predictions)
nb_report = classification_report(y_test, nb_predictions)
```
